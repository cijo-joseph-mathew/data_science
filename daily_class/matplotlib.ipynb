```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('tips')
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
sns.histplot(df['total_bill'],color='black',kde=True,edgecolor='white')
plt.title('bill distribution')
```

```python
df['day'].unique()
```

```python
sns.countplot(x='day',data=df,hue='day',legend=True)
plt.title('day count')
plt.show()
```

```python
sns.scatterplot(x='total_bill',y='tip',data=df,hue='sex')
plt.title('scatter plot od totalbill')
```

```python
sns.lineplot(x='size',y='tip',data=df,ci=None)
plt.title('line plot of tip vs size')
```

```python
sns.lineplot(x='size',y='tip',data=df,err_style='bars')
plt.title('line plot of tip vs size')
```

```python
sns.pairplot(df,hue='sex')
```

```python
corr=df.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title('correaltion matrix')
```
